# Using matplotlib.pyplot for basic data visualization:

```python
import matplotlib.pyplot as plt
```

```python
x = [1, 2, 5, 6, 9]
y = [4, 5, 7, 10, 15]

plt.title("Just a Graph")

plt.xlabel("This is the x-axis")
plt.ylabel("This is the y-axis")

plt.plot(x, y, linewidth=3)
plt.show()
```

```python
x = [1, 2, 3, 4, 5]
y = [15, 40, 55, 100, 75]

plt.title("Just a Bar Chart")

plt.xlabel("This is the x-axis")
plt.ylabel("This is the y-axis")

plt.bar(x, y, color="crimson")
plt.show()
```

```python
x = [1, 2, 3, 5, 8, 13, 21, 34, 45]
y = [10, 20, 40, 30, 20, 50, 55, 80, 70]

plt.title("Just a Scatterplot")

plt.xlabel("This is the x-axis")
plt.ylabel("This is the y-axis")

plt.scatter(x, y, color="g", marker="^", s=65)
plt.show()
```

```python
x = [1, 2, 3, 5, 8, 13, 21, 34]
y = [10, 20, 40, 30, 20, 50, 55, 80]

plt.title("Just a Connected Scatterplot")

plt.xlabel("This is the x-axis")
plt.ylabel("This is the y-axis")

plt.scatter(x, y)
plt.plot(x, y, color="y", linestyle="--")
plt.show()
```

```python
import random
```

```python
vector = []

for i in range(10):
    random_value = random.randint(0,50)
    vector.append(random_value)

plt.boxplot(vector)
plt.title("Just a Boxplot")
plt.show()
```

- A practical example: growth of Brazil's population, 1980-2016 (source: DATASUS):

```python
import pandas as pd
```

```python
data = pd.read_csv("populacao_brasileira.csv")
```

```python
data.head()
```

```python
data = open("populacao_brasileira.csv").readlines()

x = []
y = []

for i in range(len(data)):
    if i != 0:
        line = data[i].split(";")
        x.append(int(line[0]))
        y.append(int(line[1]))

plt.plot(x, y, color="k", linestyle="-.", linewidth=2)
plt.bar(x, y, color="#e4e4e4", width=0.6)
plt.title("Growth of Brazil's population, 1980-2016")
plt.xlabel("Year")
plt.ylabel("Population (Hundreds of Millions)")
plt.show()
```
